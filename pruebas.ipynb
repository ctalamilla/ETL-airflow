```python

def ingest_data (company):
    end_point = (f"{BASE_URL}?function={STOCK_FN}&symbol=GOOD&interval={INTERVAL}"f"&apikey={API_KEY}&datatype=json")
    data = requests.get(end_point).json()
    df = pd.DataFrame(data.get('Time Series (5min)')).T.reset_index().rename(columns={'index':'datetime', 
                                                                             '1. open': 'open', 
                                                                             '2. high': 'high', 
                                                                             '3. low':'low', 
                                                                             '4. close': 'close', 
                                                                             '5. volume': 'volume'})
    df['symbol'] = company
    df.to_json(f'./datalake/rawfile-{company}-{df.loc[0,["datetime"]].datetime[0:10]}.json')
    return df

```
